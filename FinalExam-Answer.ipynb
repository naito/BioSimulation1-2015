```python
import scipy, scipy.optimize, scipy.integrate
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

v = ["v"]; w = ["w"]; vw = ["v", "w"]

def V_nullcline(v):
    return v-(1./3.)*v**3

def W_nullcline(v, gamma, beta):
    return (1./gamma)*(v + beta)

def V_nullclineArray(v0=-2.5, v1=2.5):
    v = scipy.arange(v0, v1, 0.01)
    return v, V_nullcline(v)

def W_nullclineArray(gamma, beta, v0=-2.5, v1=2.5):
    v = scipy.arange(v0, v1, 0.01)
    return v, W_nullcline(v, gamma, beta)

def PlotNullclines(gamma=0.8, beta=0.7):
    V_null = V_nullclineArray()
    W_null = W_nullclineArray(gamma, beta)
    # vstar, wstar = FindFixedPoint(gamma, beta)
    pylab.plot(V_null[0], V_null[1], color='#6699ff', linewidth=1.5, linestyle="--", label="v nullcline")
    pylab.plot(W_null[0], W_null[1], color='#33cc66', linewidth=1.5, linestyle="--", label="w nullcline")
    # pylab.plot([vstar], [wstar], 'ro')

def FindFixedPoint(gamma, beta):
    def f(v, gamma, beta):
        return V_nullcline(v)-W_nullcline(v, gamma, beta)
    vstar = scipy.optimize.brentq(f, -2., 2., args=(gamma, beta))
    wstar = ((1./gamma)*(vstar+beta))
    return vstar, wstar

def dydt(y, t=50., eps=0.2, gamma=0.8, beta=0.7):
    v,w = y
    return [(1./eps) * (v - (1./3.)*v**3 - w), eps*(v - gamma*w + beta)]

def RunFitzNag(y0, t=50., eps=0.2, gamma=0.8, beta=0.7):
    tarray = scipy.arange(0.,t, float(t)/500.)
    ytraj = scipy.integrate.odeint(dydt, y0, tarray, args=(eps, gamma, beta))
    return ytraj, tarray

def Stimulate(delta, eps=0.2, gamma=0.8, beta=0.7):
    vstar, wstar = FindFixedPoint(gamma, beta)
    y0 = [vstar+delta, wstar]
    ytraj, tarray = RunFitzNag(y0, 50., eps, gamma, beta)
    pylab.plot(ytraj[:,0], ytraj[:,1], linewidth=1.5, label="I={}".format(delta))

def MultiStimulate(delta_min=0.0, delta_max=1.0, delta_step=0.1, eps=0.2, gamma=0.8, beta=0.7):
    vstar, wstar = FindFixedPoint(gamma, beta)
    for delta in scipy.arange(delta_min,delta_max+0.00001,delta_step):
        y0 = [vstar+delta, wstar]
        ytraj, tarray = RunFitzNag(y0, 20., eps, gamma, beta)
        pylab.plot(ytraj[:,0], ytraj[:,1], linewidth=1.5, label="I={}".format(delta))
    V_null = V_nullclineArray()
    W_null = W_nullclineArray(gamma, beta)
    pylab.plot(V_null[0], V_null[1], linestyle="--", linewidth=1.5, color='#6699ff', label="v nullcline")
    pylab.plot(W_null[0], W_null[1], linestyle="--", linewidth=1.5, color='#33cc66', label="w nullcline")

def TimeCourse(delta, plot = ["v","w"], eps=0.2, gamma=0.8, beta=0.7):
    vstar, wstar = FindFixedPoint(gamma, beta)
    y0 = [vstar, wstar]
    ytraj, tarray = RunFitzNag(y0, 10., eps, gamma, beta)

    for i in range(4):
        y0 = [ ytraj[-1,0] + delta, ytraj[-1,1]]
        #print y0
        _ytraj, _tarray = RunFitzNag(y0, 50., eps, gamma, beta)
        ytraj = scipy.vstack(( ytraj, _ytraj ))
        tarray = scipy.vstack(( tarray, _tarray ))

    if "v" in plot:
        pylab.plot(ytraj[:,0], label="v (I={})".format(delta))
    if "w" in plot:
        pylab.plot(ytraj[:,1], label="w (I={})".format(delta))

def MultiTimeCourse(delta_min=0.0, deltamax=0.9, delta_step=0.1, plot = ["v","w"], eps=0.2, gamma=0.8, beta=0.7):
    for delta in scipy.arange(delta_min,deltamax+0.00001,delta_step):
        TimeCourse(delta, plot , eps, gamma, beta)

def ShowPhasePlot():
    pylab.xlim([-2.5,3.5])
    pylab.xlabel("v")
    pylab.ylabel("w")
    pylab.legend(fontsize=10)
    pylab.show()

def ShowTimeCoursePlot(plot = ["v","w"]):
    pylab.xlim([0.0,2500.0])
    pylab.xlabel("time")
    if ("v" in plot) and ("w" in plot):
        pylab.ylabel("v, w")
    elif ("v" in plot):
        pylab.ylabel("v")
    elif ("w" in plot):
        pylab.ylabel("w")

    pylab.legend(fontsize=10)
    pylab.show()
```

### Question 3b 解答欄 answer column

この β, γ の組み合せでは、*v*, *w* のヌルクラインの交点（平衡点）が３つある。この平衡点のどれかに達すれば、システムは静止する。*I*ext = 0.5 では、*I*ext 印加後ただちに *v* が減少に転じ、*vx* 相空間で間近にある左下の平衡点に到達してシステムが安定する。

一方、*I*ext = 1.0 では、軌跡が *v* ヌルクラインの右側へと引きよせられた結果、右上の平衡点に到達してシステムが安定する。*I*ext による２回目の刺激からは、この平衡点の近傍を巡って同じ平衡点に戻ってくるため、時刻ゼロとは別の静止状態で、*v* の振幅が小さい反応を見せるだけになる。
